```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

store_DF = pd.read_csv('../Data/Model_Data.csv', index_col = 0)
```

```python
X = store_DF.to_numpy()
```

```python
plt.scatter(X[:,0],
            X[:,1],
            c='black',
            marker='o')
plt.grid()
plt.show()
```

### From the above plot, it is difficult to tell how many clusters there should ultimately be. An Elbow Plot is required.

```python
def plot_distortion(X,max_clusters = 10):
    distortions = []
    for i in range(1, max_clusters +1):
        km = KMeans(n_clusters=i,
                    init='k-means++',
                    n_init=10,
                    random_state=0)
        km.fit(X)
        distortions.append(km.inertia_)
        
    plt.plot(range(1,max_clusters +1), distortions, marker='o')
    plt.xlabel('Number of clusters')
    plt.ylabel('Distortion')
    plt.show() 
```

```python
# elbow rule
plot_distortion(X,max_clusters=10)
```

### Either 3 or 5 would be most appropriate.

```python
 def plot_clusters(X,y_res, plt_cluster_centers = False):
    X_centroids = []
    Y_centroids = []
    
    for cluster in set(y_res):
        x = X[y_res == cluster,0]
        y = X[y_res == cluster,1]
        X_centroids.append(np.mean(x))
        Y_centroids.append(np.mean(y))
        
        plt.scatter(x,
                    y,
                    s=50,
                    marker='s',
                    label=f'cluster {cluster}')
        
    if plt_cluster_centers:
        plt.scatter(X_centroids,
                    Y_centroids,
                    marker='*',
                    c='red',
                    s=250,
                    label='centroids')
    plt.legend()
    plt.grid()
    plt.show()
```

### KMeans with 3 clusters

```python
km_3 = KMeans(n_clusters=3, 
            n_init=10,
            random_state=0)

y_km_3 = km_3.fit_predict(X)
```

```python
 # plot clustering result
plot_clusters(X, y_km_3, plt_cluster_centers= True)
```

The three clusters above divide the data nicely. The outliers are split between Clusters 0 and 2. Cluster 1 appears to be much larger than the other two, and represents a sort of average of the total dataset. 

### KMeans with 5 clusters

```python
km_5 = KMeans(n_clusters=5, 
            n_init=10,
            random_state=0)

y_km_5 = km_5.fit_predict(X)
```

```python
 # plot clustering result
plot_clusters(X, y_km_5, plt_cluster_centers= True)
```

The results are some fair looking clusters. Cluster 4 appears to be just a collection of outliers, and that might be useful for further analysis. Clusters 0, 2, and 3 are all well defined. Cluster 1 has a few distant outliers, but still maintains the majority of data around its centroid.
